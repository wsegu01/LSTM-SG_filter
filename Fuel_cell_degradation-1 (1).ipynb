```python
data_len = len(fuelcell_dataset)
i = 0
hrs = 1

curr = []
vol = []
time = []
while(i+hrs*3600 < data_len):
  sub_data = fuelcell_dataset.iloc[i:i+hrs*3600,:]
  sub_data['current'] = sub_data['current'].astype(float)
  sub_data['voltage'] = sub_data['voltage'].astype(float)
  G = sub_data.groupby('current')['voltage'].mean()
  G = G.to_frame()
  curr = np.append(curr, G.index.to_numpy())
  vol = np.append(vol, savgol_filter(G.iloc[:,0].to_numpy(), 9, 7))
  time = np.append(time, [int(i/3600)]*len(G.index.to_numpy()))
  i += hrs*3600
```

```python
dataset = pd.DataFrame([time, curr, vol], index = ['Time', 'current', 'voltage']).T
dataset
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(dataset[['Time','current']], dataset['voltage'], test_size=0.2, shuffle=False)
```

```python
X_train, X_dev, y_train, y_dev = train_test_split(X_train, y_train, test_size=0.2, shuffle=False)
```

```python
# from sklearn.preprocessing import StandardScaler
# # from sklearn.preprocessing import scale
# scaler = StandardScaler()
# X_train['current'] = scaler.fit_transform(np.expand_dims(X_train['current'].to_numpy(), axis = 1))
X_train['current']  = X_train['current']  /X_train['current'] .abs().max()
X_train
```

```python
X_dev['current'] = X_dev['current']/X_dev['current'].abs().max()
```

```python
import tensorflow as tf
from tensorflow.keras import layers
# tf.random.set_seed(42)

inputs = layers.Input(shape = (2))
x =layers.Lambda(lambda x: tf.expand_dims(x, axis=1))(inputs)
# x = layers.LSTM(128, activation = "relu", return_sequences = True)(x)
x = layers.LSTM(64, activation = "relu")(x)
# x = layers.Dense(64, activation = 'relu')(x)
output = layers.Dense(1, activation = 'linear')(x)
model = tf.keras.Model(inputs = inputs, outputs = output, name = "model_LSTM")

# Compile
model.compile(loss = "mape",
              optimizer = tf.keras.optimizers.Adam())

# Fit
model.fit(  X_train,
            y_train,
            epochs = 100,
            verbose = 1,
            batch_size = 32,
            validation_data = (X_dev, y_dev)
            )
```

```python
X_test['current'] = X_test['current']/X_test['current'].abs().max()
predictions = model.predict(X_test)
```

```python
plt.figure(figsize = (20,7))
plt.plot(y_test.to_numpy())
plt.plot(predictions)
plt.xlabel("Time")
plt.ylabel("Voltage (V)")
plt.title("Voltage degradation curve of test dataset (767 hrs to 958 hrs)")
```
